```
# EV_Adoption_India_Analysis.ipynb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("ev_adoption_india.csv")

# Preview
print(df.head())
print(df.info())
print(df.describe())
```

## Exploratory Data Analysis (EDA)

In this section, we explore the dataset to identify patterns and trends.  

We begin by visualizing total EV sales across years to understand **overall market growth.**

Subsequent sections will break this down further into state-level and category-wise trends.


```
yearly_sales = df.groupby("Year")["EV_Sales"].sum().reset_index()

plt.figure(figsize=(8,5))
sns.lineplot(data=yearly_sales, x="Year", y="EV_Sales", marker="o")
plt.title("EV Adoption Growth in India (2019â€“2024)")
plt.xlabel("Year")
plt.ylabel("Total EV Sales")
plt.grid(True)
plt.show()
```

###  Top States Leading EV Adoption

This analysis highlights which Indian states have recorded the highest cumulative EV sales from 2019 to 2024.  
It helps us understand regional adoption patterns and which states are driving Indiaâ€™s EV revolution.  
The visualization below uses a bar chart to **compare total EV sales across states.**



```
top_states = df.groupby("State")["EV_Sales"].sum().reset_index().sort_values(by="EV_Sales", ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(data=top_states, x="EV_Sales", y="State", palette="viridis")
plt.title("Top Indian States by Total EV Adoption (2019â€“2024)")
plt.xlabel("EV Sales")
plt.ylabel("State")
plt.show()
```

## âš¡ Category-wise Growth Trends

This section visualizes how different EV categories (2W, 3W, 4W, and Buses) have evolved over time.
  
It gives a clear picture of **which segment is growing fastest**and how consumer preferences are shifting.


```
plt.figure(figsize=(8,5))
sns.lineplot(data=df, x="Year", y="EV_Sales", hue="Category", marker="o")
plt.title("Category-wise EV Growth in India (2W, 3W, 4W, Bus)")
plt.xlabel("Year")
plt.ylabel("EV Sales")
plt.grid(True)
plt.show()
```

### Key Insights
- EV sales in India have grown steadily from 2019 to 2024.
- 2W and 3W dominate the EV market, especially in Maharashtra, Delhi, and Karnataka.
- Southern and Western states lead adoption, while some northern states show late but rapid growth.
